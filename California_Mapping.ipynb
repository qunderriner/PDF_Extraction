```python
import pandas as pd
import numpy as np
import re
pd.set_option("display.max_rows", None)
```

```python
df = pd.read_csv("all_extracted_files (1).csv")
df2 = pd.read_csv("all_extracted_files_perm.csv")
```

```python
df2 = df2.rename(columns={"Term \n(months)":"Term (months)",
                          "Interest \nRate":"Interest Rate",
                          "Residual \nReceipts / \nDeferred Pymt.":
                          "Residual Receipts/Deferred Pymt.",
                          "Annual Debt \nService":"Annual Debt Service",
                          "Amount of \nFunds":"Amount of Funds"})
```

```python
#create markers for financing type 
df["Financing Type"] = "Construction Financing"
df2["Financing Type"] = "Permanent Financing"
```

```python
#in the scraping of perm finacing there seem to be two rival Interest rate columns, need to combine them 
df2["Interes\nt Rate"] = df2["Interes\nt Rate"].fillna('')#need to make NaNs blank strings so can concat columns, 
df2["Interest Rate"] = df2["Interest Rate"].fillna('')# can do this because they don't seem to contradict each other
df2["Interest Rates"] = df2["Interest Rate"] + df2["Interes\nt Rate"]
df2 = df2.drop(["Interes\nt Rate","Interest Rate"],axis=1)
df2 = df2.rename(columns={"Interest Rates":"Interest Rate"})

#note - will need to do pattern based regex to drop wrong types of interest rates and change type 
#before can do any analysis of them 
```

```python
#combine dataframes 
df_combined = pd.concat([df,df2])
```

```python
#drop GP and LP equity per email from Carolina 
df_combined = df_combined[~df_combined["Source"].str.contains("GP")]
df_combined = df_combined[~df_combined["Source"].str.contains("LP")]
```

```python
def label_financing_type(x):
    if "developer fee" in x.lower():
        return "Deferred Developer Fee"
    if "bank" in x.lower():
        return "Bank Loan"
    if "city" in x.lower():
        return "Local Grant"
    if "california" in x.lower():
        return "State Grant"
    if "donated land" in x.lower():
        return "Donated Land"
    if "boston" in x.lower():
        return "Bank Loan"
    if "county" in x.lower():
        return "Local Grant"
    if "usda" in x.lower():
        return "Federal Loan"
    if "fhlb" in x.lower():
        return "Federal Home Loan Bank"
    if "sponsor loan" in x.lower():
        return "Local Grant"
    
```

```python
df_combined["Sourcing Type"] = df_combined["Source"].apply(lambda x: label_financing_type(x))
```
